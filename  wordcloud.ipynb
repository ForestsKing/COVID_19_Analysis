```python
import wordcloud
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
```

### 以 Year 为分析粒度

#### 中文数据集

```python
# 计算TF-IDF权重
df = pd.read_csv('./data/Weibo.csv')
tfidf = TfidfVectorizer(max_features=5000)
weights = tfidf.fit_transform([' '.join(df['content'].values.tolist())]).toarray()
words = tfidf.get_feature_names_out()

# 保存权重
df = pd.DataFrame(weights, columns=words)
df.to_csv('./result/wordcloud/2020/ch/weight.csv', index=False)

# 提取权重
word2weight = {}
for word, weight in zip(words, weights[0]):
    word2weight[word] = weight

# 绘制云图
wc = wordcloud.WordCloud(
    font_path="/System/Library/fonts/PingFang.ttc",
    width=800,
    height=400,
    background_color="white"
)
wc.fit_words(word2weight)
wc.to_file("./result/wordcloud/2020/ch/wordcloud.png")
plt.imshow(wc)
plt.axis('off')
plt.show()
```

### 以 Month 为分析粒度

```python
months = ['202001', '202002', '202003', '202004', '202005', '202006', 
          '202007', '202008', '202009', '202010', '202011', '202012',]
```

#### 中文数据集

```python
# 计算TF-IDF权重
df = pd.read_csv('./data/Weibo.csv')
content = []
for month, tmp_df in df.groupby('month'):
    content.append(' '.join(tmp_df['content'].values.tolist()))
tfidf = TfidfVectorizer(max_features=5000)
weights = tfidf.fit_transform(content).toarray()
words = tfidf.get_feature_names_out()

# 保存权重
df = pd.DataFrame(weights, columns=words)
df.to_csv('./result/wordcloud/2020/ch/weight_month.csv', index=False)


for i in range(12):
    month = months[i]
    print(month)
    
    # 提取权重
    word2weight = {}
    for word, weight in zip(words, weights[i]):
        word2weight[word] = weight

    # 绘制云图
    wc = wordcloud.WordCloud(
        font_path="/System/Library/fonts/PingFang.ttc",
        width=800,
        height=400,
        background_color="white"
    )
    wc.fit_words(word2weight)
    wc.to_file("./result/wordcloud/"+month+"/ch/wordcloud.png")
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```

#### 英文数据集

```python
# 计算TF-IDF权重
df = pd.read_csv('./data/Twitter.csv')
content = []
for month, tmp_df in df.groupby('month'):
    content.append(' '.join(tmp_df['content'].values.tolist()))
tfidf = TfidfVectorizer(max_features=5000)
weights = tfidf.fit_transform(content).toarray()
words = tfidf.get_feature_names_out()

# 保存权重
df = pd.DataFrame(weights, columns=words)
df.to_csv('./result/wordcloud/2020/en/weight_month.csv', index=False)

for i in range(12):
    month = months[i]
    print(month)
    
    # 提取权重
    word2weight = {}
    for word, weight in zip(words, weights[i]):
        word2weight[word] = weight

    # 绘制云图
    wc = wordcloud.WordCloud(
        font_path="/System/Library/fonts/PingFang.ttc",
        width=800,
        height=400,
        background_color="white"
    )
    wc.fit_words(word2weight)
    wc.to_file("./result/wordcloud/"+month+"/en/wordcloud.png")
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```
